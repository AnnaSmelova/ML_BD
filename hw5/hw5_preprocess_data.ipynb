**Считаем датасет train.csv, удалим в признаке comment_text запятые и \n и сохраним его в train_prep.csv**

```python
import pandas as pd
```

```python
df = pd.read_csv('data/train.csv')
df.head(10)
```

```python
df['comment_text'].to_list()[3]
```

```python
df.shape
```

```python
def preprocess_text(text):
    text = text.replace('\n', '')
    text = text.replace(',', '')
    return text

df_prep = df.copy()
texts = df_prep['comment_text'].to_list()
texts_prep = list(map(lambda x: preprocess_text(x), texts))
df_prep['comment_text'] = texts_prep
df_prep.head(10)
```

```python
df_prep.to_csv('data/train_prep.csv', index=False)
```
